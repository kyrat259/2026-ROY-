```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold,RandomizedSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
from nba_api.stats.static import players
from nba_api.stats.endpoints import playercareerstats
from nba_api.stats.endpoints import drafthistory, leaguestandings, leaguedashteamstats,leaguedashplayerstats
import time
```

## Data Selection

Acquired data from the NBA API 
- https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/drafthistory.md

```python
draft = drafthistory.DraftHistory().get_data_frames()[0]
draft = draft[['PERSON_ID', 'SEASON', 'ROUND_NUMBER', 'OVERALL_PICK', 'ORGANIZATION_TYPE']]
draft.rename(columns={'SEASON': 'DRAFT_YEAR'}, inplace=True)
draft_df = draft[(draft['DRAFT_YEAR']).astype(int) >= 2003]
```

Acquired data from the NBA API using tables such as League Dash Player Stats and League Standings 
- https://github.com/swar/nba_api/blob/master/src/nba_api/stats/endpoints/leaguedashplayerstats.py
- https://github.com/swar/nba_api/blob/master/src/nba_api/stats/endpoints/leaguestandings.py

```python
rookie_stats = []
for year in range(2003, 2025):
    season_str = f"{year}-{str(year+1)[2:]}"
    season_stats = leaguedashplayerstats.LeagueDashPlayerStats(
        season=season_str,
        player_experience_nullable='Rookie',
        timeout=120).get_data_frames()[0]
    season_stats['SEASON_YEAR'] = year
    rookie_info = pd.merge(
        season_stats,
        draft,                  
        left_on='PLAYER_ID',
        right_on='PERSON_ID',
        how='left')
    team_stats = leaguestandings.LeagueStandings(
        season=season_str,
        timeout=120).get_data_frames()[0]
    team_stats = team_stats[['TeamID', 'WINS', 'LOSSES', 'WinPCT']]
    rookies = pd.merge(
        rookie_info,
        team_stats,
        left_on='TEAM_ID',
        right_on='TeamID',
        how='left')
    rookie_stats.append(rookies)
    time.sleep(0.7)
all_rookies = pd.concat(rookie_stats, ignore_index=True)
all_rookies['OVERALL_PICK'] = all_rookies['OVERALL_PICK'].fillna(60)
all_rookies['ROUND_NUMBER'] = all_rookies['ROUND_NUMBER'].fillna(2)
all_rookies['ORGANIZATION_TYPE'] = all_rookies['ORGANIZATION_TYPE'].fillna('Undrafted')
```

Acquired data from Basketball Reference to get a table of every ROY winner (Name + Season) and used the NBA API to match players who have won with there Id.
- https://www.basketball-reference.com/awards/roy.html
- https://github.com/swar/nba_api/blob/master/src/nba_api/stats/endpoints/playercareerstats.py

```python
url = "https://www.basketball-reference.com/awards/roy.html"
roy_data = pd.read_html(url)[0]  

if isinstance(roy_data.columns, pd.MultiIndex):
    roy_data.columns = ['_'.join([str(i) for i in col if i]).strip() for col in roy_data.columns]
roy_data = roy_data[['Unnamed: 0_level_0_Season', 'Unnamed: 2_level_0_Player']].copy()
roy_data.rename(columns={
    'Unnamed: 2_level_0_Player': 'ROY',
    'Unnamed: 0_level_0_Season': 'Season'}, inplace=True)
roy_data['SEASON_YEAR'] = (roy_data['Season'].str[:4 ].astype(int))
roy_winners = dict(zip(roy_data['SEASON_YEAR'], roy_data['ROY']))

all_players = players.get_players()
def get_id(name):
    match = players.find_players_by_full_name(name)
    return match[0]['id'] if match else None
roy_data['PLAYER_ID'] = roy_data['ROY'].apply(get_id)

roy_winners = dict(zip(roy_data['SEASON_YEAR'], roy_data['PLAYER_ID']))

all_rookies['ROY_WIN'] = 0

for year, id in roy_winners.items():
    all_rookies.loc[
        (all_rookies['SEASON_YEAR'] == year) &
        (all_rookies['PLAYER_ID'] == id),
        'ROY_WIN'] = 1
```

### Selecting Features

```python
n_features = ['PTS','PTS_RANK','REB','AST','STL','BLK','FG_PCT','FT_PCT','TOV','WINS','LOSSES','OVERALL_PICK','ROUND_NUMBER','WinPCT']
cat_feature = ['ORGANIZATION_TYPE']

X_num = all_rookies[n_features]
X_cat = pd.get_dummies(all_rookies[cat_feature], drop_first=True)
X = pd.concat([X_num, X_cat], axis=1)
y = all_rookies['ROY_WIN']
groups = all_rookies['SEASON_YEAR']
```
